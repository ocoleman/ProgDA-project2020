# Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')
pd.set_option('display.max_rows', 500)
```

```python
sleep1 = np.random.normal(334,110,500)
sleep2 = np.random.normal(302,99,500)
sleep3 = np.random.normal(290,88,500)
sleep4 = np.random.normal(254,78,500)
sleepTotal = np.concatenate((sleep1, sleep2, sleep3, sleep4), axis=None)

df = pd.DataFrame(sleepTotal, columns=["Sleep Amount (mins)"])
grades = np.random.normal(70.68,11.01,2000)
grades[::-1].sort()
df["Grade"] = grades
```

```python
for i in df.index:
    if df.at[i, 'Grade'] < 69.4:
        df.at[i, 'Grade Range'] = "<69.4%"
    if df.at[i, 'Grade'] > 69.4:
        df.at[i, 'Grade Range'] = "69.5-79.4%"
    if df.at[i, 'Grade'] > 79.4:
        df.at[i, 'Grade Range'] = "79.5-89.4%"
    if df.at[i, 'Grade'] > 89.4:
        df.at[i, 'Grade Range'] = ">89.5%"
    if df.at[i, 'Grade'] > 100.00:
        df.at[i, 'Grade'] = 100.00

for i in df.index:
    if (df.at[i, 'Grade Range'] == ">89.5%") & (df.at[i, 'Sleep Amount (mins)'] > np.random.normal(395,81)):
        df.at[i, 'Less Sleep Than Normal'] = "No"
    elif (df.at[i, 'Grade Range'] == "79.5-89.4%") & (df.at[i, 'Sleep Amount (mins)'] > np.random.normal(379,67)):
        df.at[i, 'Less Sleep Than Normal'] = "No"
    elif (df.at[i, 'Grade Range'] == "69.5-79.4%") & (df.at[i, 'Sleep Amount (mins)'] > np.random.normal(377,65)):
        df.at[i, 'Less Sleep Than Normal'] = "No"
    elif (df.at[i, 'Grade Range'] == "<69.4%") & (df.at[i, 'Sleep Amount (mins)'] > np.random.normal(367,67)):
        df.at[i, 'Less Sleep Than Normal'] = "No"
    else:
        df.at[i, 'Less Sleep Than Normal'] = "Yes"

```
