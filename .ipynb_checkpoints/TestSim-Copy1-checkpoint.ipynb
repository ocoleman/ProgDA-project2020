```python
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')
pd.set_option('display.max_rows', 500)
```

```python
Sleep = np.round(np.random.normal(1.9, 1, 200), 2)
df = pd.DataFrame({'Sleep_Time': Sleep})
df["Sleep_Time"] = pd.to_datetime(df.Sleep_Time, unit='h').dt.strftime('%H:%M:%S')

Wake = np.round(np.random.normal(9.28, 0.9, 200), 2)
df["Wake_Time"] = Wake
df["Wake_Time"] = pd.to_datetime(df.Wake_Time, unit='h').dt.strftime('%H:%M:%S')

df.Sleep_Time = pd.to_timedelta(df.Sleep_Time)
df.Wake_Time = pd.to_timedelta(df.Wake_Time)
duration = abs(df.Wake_Time - df.Sleep_Time)
df["Sleep_Duration"] = duration
```

```python
medianWakeTime = datetime.timedelta(hours=9.2)
medianSleepTime = datetime.timedelta(hours=1.78)

for i in df.index:
    if df.at[i, 'Sleep_Duration'] > datetime.timedelta(hours=12.5):
        df.at[i, 'Sleep_Duration'] = df.at[i, 'Wake_Time'] + (datetime.timedelta(hours=24) - df.at[i, 'Sleep_Time'])
    if df.at[i, 'Sleep_Time'] < medianSleepTime:
        df.at[i, 'Grade_Score'] = np.random.normal(77.25,13.71)
        df.at[i, 'Before_SleepTime'] = "Yes"
    if df.at[i, 'Sleep_Time'] > medianSleepTime:
        df.at[i, 'Grade_Score'] = np.random.normal(70.68,11.01)
        df.at[i, 'Before_SleepTime'] = "No"
    if df.at[i, 'Sleep_Time'] > datetime.timedelta(hours=21):
        df.at[i, 'Grade_Score'] = np.random.normal(77.25,13.71)
        df.at[i, 'Before_SleepTime'] = "Yes"
    if df.at[i, 'Wake_Time'] < medianWakeTime:
        df.at[i, 'Before_WakeTime'] = "Yes"
        df.at[i, 'Grade_Score'] = np.random.normal(78.28,9.33)
    if df.at[i, 'Wake_Time'] > medianWakeTime:
        df.at[i, 'Before_WakeTime'] = "No"
        df.at[i, 'Grade_Score'] = np.random.normal(69.63,14.38)
        

for i in df.index:
    if df.at[i, 'Grade_Score'] > 100:
        df.at[i, 'Grade_Score'] = 100

df
```

```python
length = np.array(df["Sleep_Duration"] / np.timedelta64(1, 'h'))
np.round(length,2)

sns.regplot(data=df, x=length, y=df["Grade_Score"])
```

```python
sleep = np.array(df["Sleep_Time"] / np.timedelta64(1, 'h'))
np.round(sleep,2)
for n, i in enumerate(sleep):
    if i > 21:
        sleep[n] -= 24

sns.scatterplot(data=df, x=sleep, y="Grade_Score", hue="Before_SleepTime")
```

```python
wake = np.array(df["Wake_Time"] / np.timedelta64(1, 'h'))
np.round(wake,2)

sns.scatterplot(data=df, x=wake, y="Grade_Score", hue="Before_WakeTime")
```
