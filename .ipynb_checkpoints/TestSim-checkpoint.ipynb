```python
#Median Wake Up time (9.12am)
medianWakeTime = datetime.timedelta(hours=9.2)
#Median Sleep Time (1.47am)
medianSleepTime = datetime.timedelta(hours=1.78)

#Loop through dataframe
for i in df.index:
    #check if sleep is abnormally high (went to bed before midnight) and correct sleep duration.
    if df.at[i, 'Sleep_Duration'] > datetime.timedelta(hours=12.5):
        df.at[i, 'Sleep_Duration'] = df.at[i, 'Wake_Time'] + (datetime.timedelta(hours=24) - df.at[i, 'Sleep_Time'])
    #If Sleep time was before the median, slightly better performance expected
    if df.at[i, 'Sleep_Time'] < medianSleepTime:
        df.at[i, 'Grade_Score'] = np.random.normal(77.25,13.71)
        df.at[i, 'Before_SleepTime'] = "Yes"
    #If sleep time was after the median, slightly worse performance expected
    if df.at[i, 'Sleep_Time'] > medianSleepTime:
        df.at[i, 'Grade_Score'] = np.random.normal(70.68,11.01)
        df.at[i, 'Before_SleepTime'] = "No"
    #If sleep time was before midnight, higher performance expected
    if df.at[i, 'Sleep_Time'] > datetime.timedelta(hours=21):
        df.at[i, 'Grade_Score'] = np.random.normal(77.25,13.71)
        df.at[i, 'Before_SleepTime'] = "Yes"

#Accounting for grades above 100.
for i in df.index:
    if df.at[i, 'Grade_Score'] > 100:
        df.at[i, 'Grade_Score'] = 100   


#Converting sleep time to a float and replacing values before midnight.
sleep = np.array(df["Sleep_Time"] / np.timedelta64(1, 'h'))
np.round(sleep,2)
for n, i in enumerate(sleep):
    if i > 21:
        sleep[n] -= 24

#Fitting line to data
m, c = np.polyfit(sleep, df["Grade_Score"], 1)
m, c       
      
#Scatterplots: Sleep Time vs Grade
fig, axes = plt.subplots(3,1, figsize=(12,20))
sns.scatterplot(ax=axes[0], data=df, x=sleep, y="Grade_Score", hue="Before_SleepTime")
axes[0].plot(sleep, m * sleep + c, 'r-')
axes[0].set_xlabel('Sleep Time(am)')
axes[0].set_ylabel('Grade Score %')
axes[0].set_title('Sleep Time vs Grade Score')


#Loop thorugh Dataframe again
for i in df.index:
    #If wake time is less than the median, significant performance increase
    if df.at[i, 'Wake_Time'] < medianWakeTime:
        df.at[i, 'Grade_Score'] = np.random.normal(78.28,9.33)
        df.at[i, 'Before_WakeTime'] = "Yes"
    #If wake time after, significant decrease to performance.
    if df.at[i, 'Wake_Time'] > medianWakeTime:
        df.at[i, 'Grade_Score'] = np.random.normal(69.63,14.38)
        df.at[i, 'Before_WakeTime'] = "No"

#Accounting for grades above 100.
for i in df.index:
    if df.at[i, 'Grade_Score'] > 100:
        df.at[i, 'Grade_Score'] = 100
        
#Converting wake time to a float   
wake = np.array(df["Wake_Time"] / np.timedelta64(1, 'h'))    
#Fitting a line to the data
m2, c2 = np.polyfit(wake, df["Grade_Score"], 1)
m2, c2

#Scatterplot: Wake Time vs Grade Score
sns.scatterplot(ax=axes[1], data=df, x=wake, y="Grade_Score", hue="Before_WakeTime")
axes[1].plot(wake, m2 * wake + c2, 'r-')
axes[1].set_xlabel('Wake Time(am)')
axes[1].set_ylabel('Grade Score %')
axes[1].set_title('Wake Time vs Grade Score')

#Converting Sleep Duration to a float 
length = np.array(df["Sleep_Duration"] / np.timedelta64(1, 'h'))
np.round(length,2)

#Fitting a line to the data
m3, c3 = np.polyfit(length, df["Grade_Score"], 1)
m3, c3

#Scatterplot: Sleep Duration vs Grade Score
sns.scatterplot(ax=axes[2],data=df,x=length,y=df["Grade_Score"])
axes[2].plot(length, m3 * length + c3, 'r-')
axes[2].set_xlabel('Sleep Hours')
axes[2].set_ylabel('Grade Score')
axes[2].set_title('Sleep Duration vs Grades')
```
